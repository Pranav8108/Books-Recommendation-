## Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

```python
popular_df['Image-URL-M'][0]
```

## Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
experiencedUsers = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(experiencedUsers)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import math

# Split data into training and testing sets
train_data, test_data = train_test_split(final_ratings, test_size=0.2)

# Create pivot table for training data
train_pt = train_data.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
train_pt.fillna(0,inplace=True)

# Calculate cosine similarity scores for training data
train_similarity_scores = cosine_similarity(train_pt)

# Create pivot table for testing data
test_pt = test_data.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
test_pt.fillna(0,inplace=True)

# Make predictions for testing data
test_predictions = []
for index, row in test_pt.iterrows():
    similarity_sum = 0
    weighted_sum = 0
    for i in range(len(train_pt)):
        if train_pt.index[i] != index:
            similarity = train_similarity_scores[i][train_pt.index.get_loc(index)]
            similarity_sum += similarity
            weighted_sum += similarity * train_pt.iloc[i][row.name]
    if similarity_sum == 0:
        prediction = 0
    else:
        prediction = weighted_sum / similarity_sum
    test_predictions.append(prediction)

# Calculate RMSE for testing data
rmse = math.sqrt(mean_squared_error(test_pt.stack(), test_predictions))
print('RMSE:', rmse)
```
